```python
#Import dependencies 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("heart.csv") 
```

```python
#plotting a bar graph
df.plot(x="age", y="trestbps", kind="bar")
```

```python
plt.hist(df['age'])
plt.xlabel("age")
```

```python
df.boxplot(column=None, by=None, ax=None, fontsize=None, rot=0, grid=True, figsize=None, layout=None)
```

```python
plt.scatter(df['age'], df['cp'])
plt.xlabel("age")
plt.ylabel("cp")
plt.show()
```

```python
#plotting a pie chart
# Plotting the pie chart for above dataframe
df.groupby(['age']).sum().plot(kind='pie', y='cp', autopct='%1.0f%%')
```
